# Steps for Shelf Transport and Shelf Eddy Analysis and Plots for TXLA Model Output

Simulations for looking at transport around the TXLA model domain and for some of the analysis for the shelf eddies were done in the `shelf_transport` project directory. Surface drifters were seeded uniformly in projected space throughout the numerical domain with 1 km spacing in both x and y. They were initialized daily from 2004 through 2010 and run for 30 days for each simulation. Subgrid diffusion was not used due to its effect on the FSLE, though an old set of simulations that did use the proper amount of subgrid diffusion for matching the mean square separation from LaCasce 2008 is available on rho.tamu.edu. The simulations used for analysis and this work are available on hafen.

### Cross-shelf connectivity

The seasonal connectivity plot was run using `make_plots.py` in the `shelf_transport` project directory. I chose January and February to represent the winter or non-summer to avoid having to span multiple years for a season (i.e., December 2004, January 2005 and February 2005). I varied different consecutive pairs of months around summer to see which showed the shelf eddy signal most strongly. Between May-June, June-July, July-August, and August-September, the July-August combination visually gave the strongest signal for the shelf eddy region, which I am taking to mean that it best represents the distinct summer dynamics.
