## Starting locations for drifters

```python
File = '/Volumes/Emmons/projects/gisr/tracks/all_f/2004-01-01T00_0gc.npz'
d = np.load(File)
xg = d['xg'][:,0]; yg = d['yg'][:,0]
d.close()

# Use only unique starting location indices since I am not using weighting anymore. Otherwise I think I will be biasing some areas.
tups = zip(xg, yg) # associate pairs of coordinates together to check for uniqueness
# from http://stackoverflow.com/questions/17350162/delete-duplicate-tuples-from-list-of-list-in-python
seen = set()
unique_tups = []
unique_tups_inds = []
for i, x in enumerate(tups):
    temp = []
    tempind = []
    for j, y in enumerate(x):
        if y not in seen:
            seen.add(y)
            temp.append(y)
            if j==0: # only add on once
                tempind.append(i)
    if not temp == []:
        unique_tups.append(temp)
        unique_tups_inds.append(tempind)
# So, out of all this, the point is: here are the indices to use for histogramming, to not use multiple points from one starting location
inds = np.asarray(unique_tups_inds)

xg = xg[inds[:,0]]; yg = yg[inds[:,0]]

xp, yp, _ = tracpy.tools.interpolate2d(xg, yg, grid, 'm_ij2xy')

figure(figsize=(11,10))
tracpy.plotting.background(grid, mers=np.arange(-100,-80,2))
plot(xp, yp, 'gs', markersize=1, mec='g')
title('Starting locations')
savefig('../figures/starting_locations.png', bbox_inches='tight', dpi=150)
```

```python
ndays = np.array([30]) #np.arange(1,31)
# shelf_depths = np.arange(50,550,50)
shelf_depths = np.array([100]) #np.array([50,100,200,300,400,500])
subplots = np.arange(1,3)
months = np.array([[1,2,12], [6,7,8]])#,7,8]]) # Winter (DJF) and Summer (JJA)
# Instead of loop through years, just use all and skip if 2011 since missing december
# years = np.arange(2004,2011) # don't have all of 2011 to use :(
names = ['Winter', 'Summer']

plotinds = np.array([3200,9800]) # indices of examples to use

# From other analysis, I know that all of the drifters in xp, yp up to ind are unique and after that are repeats
# +1 for indexing in python since we want inclusive
ind = 10701+1

## Get numerator, which depends on shelf depth and ndays ##
# Loop through number of simulations days
for nday in ndays:
#     nday = ndays[0]

    # loop through isobaths for shelf definition
    for shelf_depth in shelf_depths:
#         shelf_depth = shelf_depths[5]

#         if shelf_depth=150 or shelf_depth==250 or shelf_depth==350 or shelf_depth==450:
#             continue
        
        figname = '../figures/seasons/sample_drifters_nday' + str(nday) + '_shelfdepth' + str(shelf_depth)

        if not os.path.exists(figname + '.png'):
                
            fig = figure(figsize=(22,10))
            
            # Loop through subplots
            for i, subplot in enumerate(subplots):
                
                xpcsave = []; ypcsave = []
                
            # plot = plots[0]
                ax = fig.add_subplot(1,2,i+1)
                tracpy.plotting.background(grid, ax=ax, mers=np.arange(-100,-80,2))   
                
                # Compile together all of the track file names for this combination
                Files = []
                for month in months[i,:]:
                    Files.extend(glob.glob('/Volumes/Emmons/projects/gisr/tracks/all_f/2006-' + str(month).zfill(2) + '-*.npz'))
                    
                # Loop through Files and plot the example drifter tracks on a map
                for File in Files:
        
                    print 'Working on ', File
                    
                    d = np.load(File)
                    xg = d['xg'][:ind,:]; yg = d['yg'][:ind,:] # ind is to eliminate repeats tracks
                    xg = xg[plotinds,:]; yg = yg[plotinds,:] # pick out tracks to plot for examples
                    d.close()       
                    hind = np.isnan(xg)
                    xp, yp, _ = tracpy.tools.interpolate2d(xg, yg, grid, 'm_ij2xy')
                    xp[hind] = np.nan; yp[hind] = np.nan
                    del(xg,yg) # don't need grid info anymore
                    
                    plot(xp[0,:].T, yp[0,:].T, color='darkcyan')
                    plot(xp[1,:].T, yp[1,:].T, color='purple')
                    
                    xpc, ypc = tracpy.tools.find_final(xp, yp)
                    xpcsave.extend(xpc); ypcsave.extend(ypc)
                
                # plot initial location
                plot(xp[:,0], yp[:,0], 'go', markersize=15)
                
                # plot final locations
                plot(xpcsave, ypcsave, 'ro', markersize=8)
                
                if i==0:
                    title('Winter 2006')
                if i==1:
                    title('Summer 2006')
                
savefig('../figures/example.png', bbox_inches='tight', dpi=150)
                    
```

## Choose grid indices to represent three regions of coastline in model domain

I used ginput() to visualize try to capture a region along the coastline and about 20 km out from the coastline in three regions: Texas bays, Chenier plain, and Louisiana coastline. I changed these from projected x,y coordinates to x,y grid index coordinates, then found the rho grid points contained in each path defined by the coordinates. Then I saved the indices defining the rho grid points in these regions in a flattened format. So they give the indices along the flattened array.

```python
fig = figure(figsize=(11,10))
tracpy.plotting.background(grid, mers=np.arange(-100,-80,2))

# Texas
d = np.load('/Volumes/Emmons/projects/gisr/calcs/all_f/xy_projected_points_definining_TX_coastline.npz')
x = np.asarray(d['pts'])[:,0]; y = np.asarray(d['pts'])[:,1];
d.close()
plot(x, y, 'r', lw=5)

# Chenier plain
d = np.load('/Volumes/Emmons/projects/gisr/calcs/all_f/xy_projected_points_definining_chenier_coastline.npz')
x = np.asarray(d['pts'])[:,0]; y = np.asarray(d['pts'])[:,1];
d.close()
plot(x, y, 'darkorange', lw=5)

# Louisiana
d = np.load('/Volumes/Emmons/projects/gisr/calcs/all_f/xy_projected_points_defining_LA_coastline.npz')
x = np.asarray(d['pts'])[:,0]; y = np.asarray(d['pts'])[:,1];
d.close()
plot(x, y, 'y', lw=5)
```

## Make histograms to find connectivity between domain and coastline regions

### Texas

```python
dconn = np.load('/Volumes/Emmons/projects/gisr/calcs/all_f/flattened_rho_grid_indices_defining_TX_coastline.npz')
cinds = dconn['inds'] # indices for flattened rho array
    
# grid index pairs defining a grid cell as being in the region of interest
TX = zip(grid['X'].flat[cinds], grid['Y'].flat[cinds])

# Start from 'gc' versions of drifter tracks
# We know the drifter depth by its grid index, and get just check if the depths are both greater and less than the cross-shelf isobath

nbins = (40,40)

# From previous analysis, I know that all of the drifters in xp, yp up to ind are unique and after that are repeats
# The first listings of each point should stay consistent
rind = 10701+1

nday = 30
# ndays = np.arange(1,31)
shelf_depth = 100
# shelf_depths = np.arange(50,550,50)
loc = '/Volumes/Emmons/projects/gisr/'

# NOTE: in the future, rerun these as 40x40x30 (ndays) x 6 (shelf depths) to save everything in one file. Currently 
# I am just rerunning for what I need for the GOMRI poster.

# Loop over weeks in year and for each, average over the 8 run years
# For a certain simulation:
Files = glob.glob(loc + 'tracks/all_f/*gc.npz')

for File in Files:
# File = Files[0]

    #     if os.path.exists(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'HcrossEnd_nday30_shelfdepth100.npz'):
    #         continue
        
    print File
    
    d = np.load(File)
    # only read in unique seed locations
    xg = d['xg'][:rind+1,:]; yg = d['yg'][:rind+1,:]; tg = d['tg']
    d.close()
    
    # Find out what drifters are connected to the Texas coastline
    inTX = np.zeros(xg.shape[0], dtype='bool')
    for i in xrange(xg.shape[0]): # loop through drifters
        drifter = zip(np.ceil(xg[i,:]).astype(int), np.ceil(yg[i,:]).astype(int)) 
        # from http://stackoverflow.com/questions/16138015/python-comparing-two-lists
        # contains true for a drifter if at some time its track is in the grid indices defining the TX coast
        inTX[i] = not set(drifter).isdisjoint(TX) # checks to see if the sets are not completely disjoint (all I need is one instance!)
    
    # to projected coordinates
    hind = np.isnan(xg) # indices where nans are location in xg, yg
    xp, yp, _ = tracpy.tools.interpolate2d(xg, yg, grid, 'm_ij2xy')
    xp[hind] = np.nan; yp[hind] = np.nan # reinstitute nans into depths of trajectories
    del(xg,yg) # don't need grid index track info anymore
    
    days = (tg-tg[0])/(3600.*24)
    
    # Then make a histogram of the starting locations of the drifters that reach the Texas coastline 
    H, xe, ye = histogram2d(xp[inTX,0], yp[inTX,0], bins=nbins, \
                            range=[[grid['xr'].min(), grid['xr'].max()], \
                                    [grid['yr'].min(), grid['yr'].max()]])
        
    # Save cross histogram
    np.savez(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'H-TX-Start_nday' + str(nday) + '.npz', H=H, xe=xe, ye=ye)
```

## Probably of making it to the Texas coast

```python
ndays = np.array([30]) #np.arange(1,31)
# shelf_depths = np.arange(50,550,50)
shelf_depth = np.array([100])#([50,100,200,300,400,500])
plots = np.arange(1,3)
months = np.array([[1,2,12], [6,7,8]])#,7,8]]) # Winter (DJF) and Summer (JJA)
# Instead of loop through years, just use all and skip if 2011 since missing december
# years = np.arange(2004,2011) # don't have all of 2011 to use :(
names = ['Winter', 'Summer']
nbins = (40,40)

## Get numerator, which depends on shelf depth and ndays ##
# Loop through number of simulations days
for nday in ndays:
#     nday = ndays[0]

        
    figname = '../figures/seasons/TX_nday' + str(nday)

    if not os.path.exists(figname + '.png'):
    
        print 'Working on ', figname
            
        fig = figure(figsize=(22,10))
        
        # Loop through subplots
        for i, plot in enumerate(plots):
        # plot = plots[0]
            ax = fig.add_subplot(1,2,i+1)
            tracpy.plotting.background(grid, ax=ax, mers=np.arange(-100,-80,2))   
    
            # Compile together all of the crossing file names for this combination
            HcrossFiles = []
            for month in months[i,:]:
                HcrossFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*-TX-*' + '_nday' + str(nday) + '.npz'))

            ## Get denominator from start info ##
            # Compile together all of the starting file names
            HstartFiles = []
            for month in months[i,:]:
                HstartFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*Hstart*.npz'))
                
            # This will hold the cumulative denominator of starting info for the histogram
            Hstart = np.zeros(nbins)
            # Loop over start files and compile together starting histogram info for denominator in plotted histogram
            for HstartFile in HstartFiles:
        #         if '2004-07-22' not in HstartFile:
                d = np.load(HstartFile)
                Hstart += d['Hstart']
                xe = d['xe']; ye = d['ye'] # at cell edges
                xc = op.resize(d['xe'],0); yc = op.resize(d['ye'],0) # at cell centers
                d.close()
        
            # This will hold the cumulative denominator of starting info for the histogram
            Hcross = np.zeros(nbins)
            # Loop over start files and compile together crossing histogram info for denominator in plotted histogram
            for HcrossFile in HcrossFiles:
        #         if '2004-07-22' not in HcrossFile:
                d = np.load(HcrossFile)
                Hcross += d['H']
                d.close()
                
            ## Do the plot of the combined histogram ##
            H = (Hcross.T/Hstart.T)*100
        #     ind = np.isnan(H)
        #     H[ind] = 0

            contourf(xc, yc, H, cmap='YlGnBu', levels=np.linspace(0,100,11))
            c = ax.contour(grid['xr'], grid['yr'], grid['h'], [shelf_depth], colors='0.2', linewidths=1)
#                 clabel(c, fmt='%i', colors='0.2', fontsize=20)
            
            title(names[i])
            if i==0:
                if nday==1:
                    text(0.6, 0.01, '%i simulation day' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                else:
                    text(0.6, 0.01, '%i simulation days' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                        
        fig.suptitle('Origin of water reaching Texas coast', fontsize=30, y=.94)
        
        # Horizontal colorbar below plot
        cax = fig.add_axes([0.25, 0.05, 0.48, 0.03]) #colorbar axes
        cb = colorbar(cax=cax, orientation='horizontal')
        cb.set_label('Probability (\%)')
        # do 300dpi once I've decided which plots are worth more resolution
        savefig(figname + '.png', bbox_inches='tight', dpi=150)
#         savefig('../figures/seasons/start_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.pdf', bbox_inches='tight')
        close()
```

## Probability of making it to Louisiana coast

```python
ndays = np.array([30]) #np.arange(1,31)
# shelf_depths = np.arange(50,550,50)
shelf_depth = np.array([100])#([50,100,200,300,400,500])
plots = np.arange(1,3)
months = np.array([[1,2,12], [6,7,8]])#,7,8]]) # Winter (DJF) and Summer (JJA)
# Instead of loop through years, just use all and skip if 2011 since missing december
# years = np.arange(2004,2011) # don't have all of 2011 to use :(
names = ['Winter', 'Summer']
nbins = (40,40)

## Get numerator, which depends on shelf depth and ndays ##
# Loop through number of simulations days
for nday in ndays:
#     nday = ndays[0]

        
    figname = '../figures/seasons/LA_nday' + str(nday)

    if not os.path.exists(figname + '.png'):
    
        print 'Working on ', figname
            
        fig = figure(figsize=(22,10))
        
        # Loop through subplots
        for i, plot in enumerate(plots):
        # plot = plots[0]
            ax = fig.add_subplot(1,2,i+1)
            tracpy.plotting.background(grid, ax=ax, mers=np.arange(-100,-80,2))   
    
            # Compile together all of the crossing file names for this combination
            HcrossFiles = []
            for month in months[i,:]:
                HcrossFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*-LA-*' + '_nday' + str(nday) + '.npz'))

            ## Get denominator from start info ##
            # Compile together all of the starting file names
            HstartFiles = []
            for month in months[i,:]:
                HstartFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*Hstart*.npz'))
                
            # This will hold the cumulative denominator of starting info for the histogram
            Hstart = np.zeros(nbins)
            # Loop over start files and compile together starting histogram info for denominator in plotted histogram
            for HstartFile in HstartFiles:
        #         if '2004-07-22' not in HstartFile:
                d = np.load(HstartFile)
                Hstart += d['Hstart']
                xe = d['xe']; ye = d['ye'] # at cell edges
                xc = op.resize(d['xe'],0); yc = op.resize(d['ye'],0) # at cell centers
                d.close()
        
            # This will hold the cumulative denominator of starting info for the histogram
            Hcross = np.zeros(nbins)
            # Loop over start files and compile together crossing histogram info for denominator in plotted histogram
            for HcrossFile in HcrossFiles:
        #         if '2004-07-22' not in HcrossFile:
                d = np.load(HcrossFile)
                Hcross += d['H']
                d.close()
                
            ## Do the plot of the combined histogram ##
            H = (Hcross.T/Hstart.T)*100
        #     ind = np.isnan(H)
        #     H[ind] = 0

            contourf(xc, yc, H, cmap='YlGnBu', levels=np.linspace(0,100,11))
            c = ax.contour(grid['xr'], grid['yr'], grid['h'], [shelf_depth], colors='0.2', linewidths=1)
#                 clabel(c, fmt='%i', colors='0.2', fontsize=20)
            
            title(names[i])
            if i==0:
                if nday==1:
                    text(0.6, 0.01, '%i simulation day' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                else:
                    text(0.6, 0.01, '%i simulation days' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                        
        fig.suptitle('Origin of water reaching Louisiana coast', fontsize=30, y=.94)
        
        # Horizontal colorbar below plot
        cax = fig.add_axes([0.25, 0.05, 0.48, 0.03]) #colorbar axes
        cb = colorbar(cax=cax, orientation='horizontal')
        cb.set_label('Probability (\%)')
        # do 300dpi once I've decided which plots are worth more resolution
        savefig(figname + '.png', bbox_inches='tight', dpi=150)
#         savefig('../figures/seasons/start_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.pdf', bbox_inches='tight')
        close()
```
