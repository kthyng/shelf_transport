```python

# Loop through drifter files by month/year. If they fit the criteria, save them to plot later.
years = np.arange(2004,2012) 
pmF = []; pbF = []; ptF = []; plF = [];
pmJ = []; pbJ = []; ptJ = []; plJ = [];
for year in years: # loop through years
    
    # Load in drifter tracks by month
    Files = glob.glob(loct + str(year) + '*-02-*.nc')
    d = netCDF.MFDataset(Files, aggdim='ntrac')
    pmF.append([d.variables['lonp'][ind[0]], d.variables['latp'][ind[0]]])
    pbF.append([d.variables['lonp'][ind[1]], d.variables['latp'][ind[1]]])
    ptF.append([d.variables['lonp'][ind[2]], d.variables['latp'][ind[2]]])
    plF.append([d.variables['lonp'][ind[3]], d.variables['latp'][ind[3]]])
    d.close()
    
    Files = glob.glob(loct + str(year) + '*-07-*.nc')
    d = netCDF.MFDataset(Files, aggdim='ntrac')
    pmJ.append([d.variables['lonp'][ind[0]], d.variables['latp'][ind[0]]])
    pbJ.append([d.variables['lonp'][ind[1]], d.variables['latp'][ind[1]]])
    ptJ.append([d.variables['lonp'][ind[2]], d.variables['latp'][ind[2]]])
    plJ.append([d.variables['lonp'][ind[3]], d.variables['latp'][ind[3]]])
    d.close()
    
    # Now the lon/lat coordinates for drifter tracjectories are stored in these lists for plotting
```

```python
# Seasonal differences in trajectories for different areas of the shelf: Mississippi Delta
fig = figure(figsize=(22,10))
ax = fig.add_subplot(1,2,1)
tracpy.plotting.background(grid, mers=np.arange(-100, -80, 2))
# Plot tracks from all 8 years
xe = []; ye = [];
for i in xrange(len(pmF)):
    x, y = grid['basemap'](pmF[i][0],pmF[i][1])
    inds = np.isnan(pmF[i][0])
    x[inds] = np.nan; y[inds] = np.nan;
    plot(x.T, y.T,'0.6', linewidth=0.5)
    xet, yet = tracpy.tools.find_final(x, y)
    xe.append(xet); ye.append(yet);
plot(x[:,0], y[:,0], 'go', markersize=5)
plot(xe, ye, 'ro', markersize=4)
ax.contour(grid['xr'], grid['yr'], grid['h'], [100], colors='k', linewidth=3, zorder=5)
title("February")

ax = fig.add_subplot(1,2,2)
tracpy.plotting.background(grid, mers=np.arange(-100, -80, 2))
# Plot tracks from all 8 years
xe = []; ye = [];
for i in xrange(len(pmJ)):
    x, y = grid['basemap'](pmJ[i][0],pmJ[i][1])
    inds = np.isnan(pmJ[i][0])
    x[inds] = np.nan; y[inds] = np.nan;
    plot(x.T, y.T,'0.6', linewidth=0.5)
    xet, yet = tracpy.tools.find_final(x, y)
    xe.append(xet); ye.append(yet);
plot(x[:,0], y[:,0], 'go', markersize=5)
plot(xe, ye, 'ro', markersize=4)
ax.contour(grid['xr'], grid['yr'], grid['h'], [100], colors='k', linewidth=3, zorder=5)
title("July")

fig.suptitle('Aggregated drifters starting near Mississippi delta, 2004-2011', fontsize=22, y=.94)
fig.savefig('../figures/area1_tracks.png', bbox_inches='tight')
```
