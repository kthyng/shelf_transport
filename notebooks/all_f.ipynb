#All Forward: running drifters from through the TX-LA model domain forward in time
Kristen Thyng, October 2013

Numerical surface drifters are simulation every 8 grid cells (five at each location in the TXLA model starting daily for February and July for the years 2004-2011 and running forward for 7 and 25 days (separately for transport calculation).

A parameterization of subgrid scale motion is implemented via random displacement (`doturb=2`) and a horizontal diffusivity of $A_H=5$ m$^2$/s is used, based on a study comparing the drifter results with drifter data from the Gulf.

The [project](https://github.com/kthyng/gisr/blob/master/projects/all_f.py) is on github.

--------

Update January 2014

Numerical surface drifters have been seeded every 5000 meters throughout the TXLA numerical domain, started weekly for 2004-2011, running for 30 days. These are being run on hafen.

All analysis here is about cross-shelf transport.

##Connectivity Maps
These are modeled after the NOAA Trajectory Analysis Planner (TAP) maps (["The NOAA Trajectory Analysis Planner: TAP II"](http://response.restoration.noaa.gov/sites/default/files/TAP_paper.pdf), Chris Barker, NOAA). They have a number of useful products all addressing pertinent response and response planning questions. I want to combine some aspects of several of these together.

```python
# This is already done, on Rainier. 2004-2010

# Start from 'gc' versions of drifter tracks
# We know the drifter depth by its grid index, and get just check if the depths are both greater and less than the cross-shelf isobath

nbins = (40,40)
  
nday = 30
# ndays = np.arange(1,31)
shelf_depth = 100
# shelf_depths = np.arange(50,550,50)
loc = '/Volumes/Emmons/projects/gisr/'

# NOTE: in the future, rerun these as 40x40x30 (ndays) x 6 (shelf depths) to save everything in one file. Currently 
# I am just rerunning for what I need for the GOMRI poster.

# Loop over weeks in year and for each, average over the 8 run years
# For a certain simulation:
Files = glob.glob(loc + 'tracks/all_f/*gc.npz')
# File = Files[0]
for File in Files:
    
#     # Histogram initialization (CHANGE THIS TO ACCOMMODATE MORE ndays and shelf depths options later)
#     H = np.zeros(nbins)
    
#     if '2004' in File or '2005' in File or '2006' in File or '2007' in File:
#         continue

    if os.path.exists(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'HcrossEnd_nday30_shelfdepth100.npz'):
        continue
    
    print File

    d = np.load(File)
    xg = d['xg']; yg = d['yg']; tg = d['tg']
    d.close()
    
    # From previous analysis, I know that all of the drifters in xp, yp up to ind are unique and after that are repeats
    # The first listings of each point should stay consistent
    rind = 10701+1


#     # Use only unique starting location indices since I am not using weighting anymore. Otherwise I think I will be biasing some areas.
#     tups = zip(xg[:,0], yg[:,0]) # associate pairs of coordinates together to check for uniqueness
#     # from http://stackoverflow.com/questions/17350162/delete-duplicate-tuples-from-list-of-list-in-python
#     seen = set()
#     unique_tups = []
#     unique_tups_inds = []
#     for i, x in enumerate(tups):
#         temp = []
#         tempind = []
#         for j, y in enumerate(x):
#             if y not in seen:
#                 seen.add(y)
#                 temp.append(y)
#                 if j==0: # only add on once
#                     tempind.append(i)
#         if not temp == []:
#             unique_tups.append(temp)
#             unique_tups_inds.append(tempind)
#     # So, out of all this, the point is: here are the indices to use for histogramming, to not use multiple points from one starting location
#     inds = np.asarray(unique_tups_inds)
    
#     xg = xg[inds[:,0],:]; yg = yg[inds[:,0],:]

    xg = xg[:rind,:]; yg = yg[:rind,:]
    
#     print '... read in track info'
    
    # Calculate the depths for the drifter positions
    hind = np.isnan(xg) # indices where nans are location in xg, yg
    xg[hind] = 0; yg[hind] = 0 # need to preserve the shape of xg, yg, and have valid indices to use with grid['h']
    h = grid['h'][np.ceil(xg).astype(int), np.ceil(yg).astype(int)]
    h[hind] = np.nan # reinstitute nans into depths of trajectories
    
    xp, yp, _ = tracpy.tools.interpolate2d(xg, yg, grid, 'm_ij2xy')
    xp[hind] = np.nan; yp[hind] = np.nan
    del(xg,yg) # don't need grid info anymore
    
    days = (tg-tg[0])/(3600.*24)
    
    # Histogram of number of drifters starting in bins with weights corresponding to their initial volume transport
    # xp0, yp0, _ = tracpy.tools.interpolate2d(xg[:,0], yg[:,0], grid, 'm_ij2xy')
    # xp0[hind[:,0]] = np.nan; yp0[hind[:,0]] = np.nan
    H, xe, ye = histogram2d(xp[:,0], yp[:,0], bins=nbins, \
                            range=[[grid['xr'].min(), grid['xr'].max()], \
                                    [grid['yr'].min(), grid['yr'].max()]])
    
    # Save start histogram
    np.savez(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'Hstart.npz', Hstart=H, xe=xe, ye=ye)
    
    # loop over simulation time, 1 to 30 days
#     print 'looping through simulation time'
#     for nday in ndays:
    # nday = ndays[0]
#         print 'nday=', nday
    
#         # Histogram of number of drifters ending in bins with weights corresponding to their initial volume transport
    ind = find(nday>=days)[-1] # number of indices into time dimension this nday corresponds to
#         # xp, yp, _ = tracpy.tools.interpolate2d(xg[:,ind], yg[:,ind], grid, 'm_ij2xy')
#         # xp[hind] = np.nan; yp[hind] = np.nan
#         H, xe, ye = histogram2d(xp[:,ind], yp[:,ind], bins=nbins, 
#                                 range=[[grid['xr'].min(), grid['xr'].max()], 
#                                         [grid['yr'].min(), grid['yr'].max()]])
    
#         # Save ending histogram, one for each day in ndays
#         np.savez(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'Hend' + '_nday' + str(nday) + '.npz', Hend=H, xe=xe, ye=ye)
    
    # loop through isobaths for shelf definition
#         for shelf_depth in shelf_depths:
    # shelf_depth = shelf_depths[0]
#             print 'shelf_depth=', shelf_depth
    
    # Using drifter locations in grid coordinates, make drifter-sized array of 1's and 0's, 1 if the depth is both above and below
    # the shelf_depth during the nday length of considered simulation, and 0 if not. This is a function of nday and shelf_depth.
    # crosses = np.zeros(T0.shape) # elements will be changed to 1 if a given drifter is both above and below isobath at some point
    hsum = np.sum(h[:,:ind]>shelf_depth, axis=1) # ONLY LOOK UP THROUGH NDAY IN TIME FOR CROSSINGS
    # need to both have at least one point greater than shelf_depth but not all non-nan positions have over shelf_depth depth
    crosses = (hsum>0)*(hsum<np.sum(~hind[:,:ind], axis=1)) # bool of whether drifter crosses shelf
    
    # Then make a histogram of the starting locations of the drifters that do cross, weighted by their T0 and save. 
    # Function of nday and shelf_depth
    #             xp, yp = tracpy.tools.interpolate2d(xg[crosses,0], yg[crosses,0], grid, 'm_ij2xy')
    H, xe, ye = histogram2d(xp[crosses,0], yp[crosses,0], bins=nbins, \
                            range=[[grid['xr'].min(), grid['xr'].max()], \
                                    [grid['yr'].min(), grid['yr'].max()]])
    
    # Save cross histogram
    np.savez(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'HcrossStart_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.npz', Hcross=H, xe=xe, ye=ye)
    
    # Also make a histogram of the ending locations of the drifters that cross, weighted by their T0 and save
    # Function of nday and shelf_depth
    # THIS SHOULDN"T BE -1 SINCE DEPENDS ON NANS BEING AT THE END OF THE TRACK
    # pdb.set_trace()
    xpcrosses = xp[:,ind]; ypcrosses = yp[:,ind] # to be modified by the new few lines
#            xpcrosses = xp[:,-1]; ypcrosses = yp[:,-1] # to be modified by the new few lines # WE WANT THIS AS A FUNCTION OF NDAY
    
    # Fill in places that had nan as final location
    nans = np.where(np.isnan(xp[:,:ind])) # find where nan's start for drifter tracks (where exit domain) #ONLY UP UNTIL NDAY SIMULATION TIME
    rows, irows = np.unique(nans[0], return_index=True) # narrow down to the first instance of the nan
    iend = nans[1][irows] - 1 # gives the 2nd index (with irows as 1st index) of the first nan, so subtract 1 to get to the last location
    xpcrosses[nans[0][irows]] =  xp[nans[0][irows],nans[1][irows]-1] # fills in nan's
    H, xe, ye = histogram2d(xpcrosses[crosses], ypcrosses[crosses], bins=nbins, \
                            range=[[grid['xr'].min(), grid['xr'].max()], \
                                    [grid['yr'].min(), grid['yr'].max()]])
    # Save cross histogram
    np.savez(loc + 'calcs/all_f/' + File.split('/')[-1][:-6] + 'HcrossEnd_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.npz', Hcross=H, xe=xe, ye=ye)
    
    # Then should be able to add these pieces together to make all other plots!
```

```python
ndays = np.arange(1,31)
# shelf_depths = np.arange(50,550,50)
shelf_depths = np.array([50,100,200,300,400,500])
plots = np.arange(1,3)
months = np.array([[1,2,12], [6,7,8]])#,7,8]]) # Winter (DJF) and Summer (JJA)
# Instead of loop through years, just use all and skip if 2011 since missing december
# years = np.arange(2004,2011) # don't have all of 2011 to use :(
names = ['Winter', 'Summer']

## Get numerator, which depends on shelf depth and ndays ##
# Loop through number of simulations days
for nday in ndays:
#     nday = ndays[0]

    # loop through isobaths for shelf definition
    for shelf_depth in shelf_depths:
#         shelf_depth = shelf_depths[5]

#         if shelf_depth=150 or shelf_depth==250 or shelf_depth==350 or shelf_depth==450:
#             continue
        
        figname = '../figures/seasons/start_nday' + str(nday) + '_shelfdepth' + str(shelf_depth)

        if not os.path.exists(figname + '.png'):
        
            print 'Working on ', figname
                
            fig = figure(figsize=(22,10))
            
            # Loop through subplots
            for i, plot in enumerate(plots):
            # plot = plots[0]
                ax = fig.add_subplot(1,2,i+1)
                tracpy.plotting.background(grid, ax=ax, mers=np.arange(-100,-80,2))   
        
                # Compile together all of the crossing file names for this combination
                HcrossFiles = []
                for month in months[i,:]:
                    HcrossFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*HcrossStart_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.npz'))
    
                ## Get denominator from start info ##
                # Compile together all of the starting file names
                HstartFiles = []
                for month in months[i,:]:
                    HstartFiles.extend(glob.glob('/Volumes/Emmons/projects/gisr/calcs/all_f/*-' + str(month).zfill(2) + '-*Hstart*.npz'))
                    
                # This will hold the cumulative denominator of starting info for the histogram
                Hstart = np.zeros(nbins)
                # Loop over start files and compile together starting histogram info for denominator in plotted histogram
                for HstartFile in HstartFiles:
            #         if '2004-07-22' not in HstartFile:
                    d = np.load(HstartFile)
                    Hstart += d['Hstart']
                    xe = d['xe']; ye = d['ye'] # at cell edges
                    xc = op.resize(d['xe'],0); yc = op.resize(d['ye'],0) # at cell centers
                    d.close()
            
                # This will hold the cumulative denominator of starting info for the histogram
                Hcross = np.zeros(nbins)
                # Loop over start files and compile together crossing histogram info for denominator in plotted histogram
                for HcrossFile in HcrossFiles:
            #         if '2004-07-22' not in HcrossFile:
                    d = np.load(HcrossFile)
                    Hcross += d['Hcross']
                    d.close()
                    
                ## Do the plot of the combined histogram ##
                H = (Hcross.T/Hstart.T)*100
            #     ind = np.isnan(H)
            #     H[ind] = 0
    
                contourf(xc, yc, H, cmap='YlOrRd', levels=np.linspace(0,100,11))
                c = ax.contour(grid['xr'], grid['yr'], grid['h'], [shelf_depth], colors='0.2', linewidths=1)
#                 clabel(c, fmt='%i', colors='0.2', fontsize=20)
                
                title(names[i])
                if i==0:
                    if nday==1:
                        text(0.6, 0.01, '%i simulation day' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                    else:
                        text(0.6, 0.01, '%i simulation days' % nday, transform=ax.transAxes, color='0.4', fontsize=22)
                            
            fig.suptitle('Origin of water crossing shelf', fontsize=30, y=.94)
            
            # Horizontal colorbar below plot
            cax = fig.add_axes([0.25, 0.05, 0.48, 0.03]) #colorbar axes
            cb = colorbar(cax=cax, orientation='horizontal')
            cb.set_label('Probability (\%)')
            # do 300dpi once I've decided which plots are worth more resolution
            savefig(figname + '.png', bbox_inches='tight', dpi=150)
    #         savefig('../figures/seasons/start_nday' + str(nday) + '_shelfdepth' + str(shelf_depth) + '.pdf', bbox_inches='tight')
            close()
```
